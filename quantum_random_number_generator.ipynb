```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
from math import pi
from numpy import arccos, sqrt
from qiskit import transpile, QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator, Aer
from qiskit.visualization import plot_histogram, plot_state_city, plot_bloch_multivector
import qiskit.quantum_info as qi
from qiskit.quantum_info import Statevector
import time
```

```python
def quantumRandomNumberGeneator(bits=5):
    print(f"Thinking of a random interger from 0 to 2^{bits}...")
    time.sleep(1)

    # Initialize a quantum circuit 
    qc1 = QuantumCircuit(1) # One qubit-circuit

    # Apply Hadamard Gate
    qc1.h(0)

    # Measure the qubit
    qc1.measure_all()

    sim = Aer.get_backend('aer_simulator') # Initialize Aer simulator
    qc1 = transpile(qc1, sim) # Transpiles (?) the quantum circuit for the simuator

    random_number_binary = np.array([], dtype=int) # Array to store binary result
    conversion_array = 2**np.arange(0, bits)[::-1] # Array to convert binary back into interger

    for i in np.arange(bits): # Go through each bit, get a 1 or 0 (with equal probability)
        result = sim.run(qc1, shots=1).result() # Run the simulator once, store result
        result = int(list(result.get_counts().keys())[0]) # Extract whether a 1  or a 0 was recorded
        random_number_binary = np.append(random_number_binary, result) # Append result to binary array
    random_interger = np.dot(random_number_binary, conversion_array)
    print("Got it!")
    return random_interger
```

```python
print(quantumRandomNumberGeneator(7))
```
